```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn
seaborn.set_theme()
import glob
```

```python
dfolder = '../RM Data/'
glob.glob('*con*', root_dir=dfolder)[0]
```

```python
hb = np.genfromtxt(dfolder+glob.glob('*hb*', root_dir=dfolder)[0])
hb_t, hb_flux, hb_f_err = hb[:,0], hb[:,1], hb[:,2]
```

```python
con = np.genfromtxt(dfolder+glob.glob('*con*', root_dir=dfolder)[0])
con_t, con_flux, con_f_err = con[:,0], con[:,1], con[:,2]
hb = np.genfromtxt(dfolder+glob.glob('*hb*', root_dir=dfolder)[0])
hb_t, hb_flux, hb_f_err = hb[:,0], hb[:,1], hb[:,2]
fig, ax = plt.subplots(2, 1, figsize=(10,7), sharex=True)
plt.subplots_adjust(hspace=0.1)
ax[0].errorbar(con_t, con_flux, yerr=con_f_err, capsize=2, fmt='.', label='Continuum')
ax[0].set_title('Continuum')
ax[0].set_ylabel('Flux')
ax[0].legend()

ax[1].errorbar(hb_t, hb_flux, yerr=hb_f_err, capsize=2, fmt='.', c='orange', label=r'H$\beta$')
ax[1].set_xlabel('time (days)')
ax[1].set_ylabel('Flux')
ax[1].legend() ;
```

```python
conrec = np.genfromtxt(r'//wsl$/Ubuntu-22.04/home/alice/brains/BRAINS/data/con_rec.txt')
linrec = np.genfromtxt(r'//wsl$/Ubuntu-22.04/home/alice/brains/BRAINS/data/line_rec.txt')
conrec_t, conrec_f = conrec[:len(conrec)//2, 0], conrec[:len(conrec)//2, 1]
conrec_t2, conrec_f2 = conrec[len(conrec)//2:, 0], conrec[len(conrec)//2:, 1]

linrec_t, linrec_f = linrec[:len(linrec)//2, 0], linrec[:len(linrec)//2, 1]
linrec_t2, linrec_f2 = linrec[len(linrec)//2:, 0], linrec[len(linrec)//2:, 1]
fig, ax = plt.subplots(2, 1, figsize=(10,7), sharex=True)
plt.subplots_adjust(hspace=0.1)
ax[0].plot(conrec_t, conrec_f, label='Continuum Reconstruction')
ax[0].plot(conrec_t2, conrec_f2, c='green', label='Continuum Reconstruction 2')
ax[0].set_title('Continuum')
ax[0].set_ylabel('Flux')
ax[0].legend()
ax[1].plot(linrec_t, linrec_f, c='orange', label=r'H$\beta$ Reconstruction')
ax[1].plot(linrec_t2, linrec_f2, c='red', label=r'H$\beta$ Reconstruction 2')
ax[1].set_xlabel('time (days)')
ax[1].set_ylabel('Flux')
ax[1].legend() ;
```
